```python
#  list  operations
# Concatenation

ls1 = [1,2,3,4,5]
ls2 = [5,6,7,8,9]
ls3 = ls1 + ls2
print("ls1  : ",ls1," len : ",len(ls1))
print("ls2  : ",ls2," len : ",len(ls2))
print("ls3  : ",ls3," len : ",len(ls3))

print("Concatenation done")

# Repetition
ls1 = [1,2,3,4,5]
ls3 = ls1 * 2
print("ls1  : ",ls1," len : ",len(ls1))
print("ls3  : ",ls3," len : ",len(ls3))
print("repetition done")
```

```python
#Most Common words in a text

import string

fhand = open('test.txt')

d = dict()

for line in fhand:
    line = line.translate(str.maketrans('','',string.punctuation))
    line = line.lower()
    
    for word in line.split():
        if word not in d:
            d[word] = 1
        else:
            d[word]+= 1
            
    ls = list()
    
    for  key,val in list(d.items()):
        ls.append((val,key))
        
    ls.sort(reverse=True)
    for key,val  in ls[:10]:
        print(key,val)
            
           
```

```python
#Tuples as keys in Dictonaries

names = (('tom','cat'),('jerry','mouse'),('donald','duck'))
number = [9686,2827,8105]

telDir={}

for i in range(len(number)):
    telDir[names[i]] = number[i]
    
for fn,ln in telDir:
    print(fn,ln,telDir[fn,ln])
```
